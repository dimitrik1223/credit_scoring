### Problem Statement: 

Determining what covariates or features have the largest impact on the likelihood of survival of a loan or conversely the likelihood of default is a useful practice, in order to both identify future lenders who are not likely to default, but also to determine which out of the current loaners may end up defaulting and predict the time to this event. One of the most popular approaches to this kind of problem is survival analysis. 

Throughout this notebook survival analysis methods from the `lifelines` library will be applied on a dataset of mortgage loans, which at any time can reflect one of three states, no observable event occurred at time *t*, defaulted at time *T*, or the loaner paid off the mortgage loan at time *t*. The dataset also contains a number of features which fluctuate over time and may prove useful in determining the probability of survival (probability of not defaulting) at any given time during the 60 month span of this data. 

The general structure of this workflow is indicated in the cell below.

```python
def miss_ratio(df):
    """
    Returns list of tuples where the first element is the column and the second is the ratio of missing values for that column within the input dataframe. 
    Args: dataframe
    """
    cols_w_nas = list(df.columns[df.isnull().any()])
    ratio_missing = [df[col].isna().sum()/len(df) for col in cols_w_nas]
    zipped_lists = zip(cols_w_nas, ratio_missing) 
    cols_missing_ratios = list(zipped_lists)
    return(cols_missing_ratios)
```

### Kaplan-Meier Estimation 

Kaplan Meier is a nonparametric model that allows for estimation of the survival function. The survival function is defined as: $S(t) = Pr(T > t)$ -- this equation is interpreted as so, the survival function at time *t* equals the probability the time at which the event of interest occurred (default) is greater than *t*. This probability is deduced using the cumulative density function of default events over time. Therefore, while plotting the survival function is useful for getting an idea of the average or aggregate survival function of a population as well as determining whether certain discrete groups or samples of the population have differing survival functions when compared, it is virtually an EDA technique. That being said, Kaplan Meire is a simple but very useful statistical test for determining if we can reject the null hypothesis that there is no difference in the survival functions of discrete groups we'd like to compare. 

* **Population-wide Kaplan Meier survival function estimation:**

The cell block below displays a survival function plot for the full dataset. As suspected, given the low incidence rate of default, the global probability of survival drops only as low as roughly 93% at the 5-year timeline endpoint. Additionally, we can see at which point the majority of defaults occured by examining the *at risk counts* at the bottom of the figure. It's visible that between 30-40 is the period with the most defaults, with 7,906 occurring within this period. That makes up just over 50% of all defaults observed within the data-collection timeline. 

* **Observing survival functions between discrete groups:** 

In order to determine which categorical groups have the most distinct variances in their Kaplan Meier survival function estimations, the `kpm_groups_auto_plot` function prototype was defined. This function intends to be a one-stop-shop for Kaplan Meire exploratory analysis; allowing a user to simply input a feature of interest, the number of groups to compare, which specific groups to consider, and optionally for categorical features, the names of the different levels to be compared. Additionally, the function supports automatic 10-bin discretization if the input feature is continuous. 

I invoke the `kpm_groups_auto_plot` within a for loop in order to quickly get a view of the survival function plots of all the features present in the `mortgage_df`. At a glance it's clear that given the short timeline and low incidence rate, it's not possible to reject the Kaplan Meier test's null hypothesis that the groups survival functions are not different from one another over time. A good indicator of this assumption is when the survival function curves of the different groups intersect one another. However, a couple of the features do offer some interesting plots, most notably the `balance_time` feature. 

* `balance_time`: When comparing the first, third, fifth, and tenth quantiles out of ten total bins, the probability of survival as a function of time, shows some rather distinct differences, although some of the curves do criss-cross one another. The first quantile which corresponds to the bucket of mortgages with the lowest outstanding balances maintains the highest probability of survival throughout the timeline, degrading only to a value of approximately .96 at 60 months. This makes logical sense as those mortgages which have the lowest balance from the beginning should be the easiest to repay, and thus the likelihood of default or hazard rate is lesser for this group. For the 3rd or 30% quantile and the 9th and highest outstanding balance quantile or bucket the survival curves stick together hardly showing any gap. However, the 9th quantile appears to remain slightly above the 3rd quantile in the end, stooping to a probability of survival of approximately  .94 or 94%. While there seems to be no clear hypothesis for why the 3rd quantile would significantly differ from the 1st quantile, it does seem evident that those loans with the highest outstanding balance at origin time would be the most likely to be defaulted on. Finally, the fifth quantile, which can be regarded as the median or midrange bin of outstanding balances, shows the fastest degradation with respect to probability of survival. The survival functions quite quickly descends to a probability of survival of around .92 -- the lowest final probability of survival for the time window.

```python
# Population-wide Survival function plot 

kmf = KaplanMeierFitter()
sns.set(rc={'figure.figsize':(12,8)})
kmf.fit(mortgage_df['time'],mortgage_df['default_time'])
kmf.plot_survival_function()
plt.ylabel('Probability of survival (Probability of not defaulting)')
plt.xlabel('Months')


add_at_risk_counts(kmf)
plt.title('Survival function of population')
plt.tight_layout()
```

```python
def kpm_groups_auto_plot(df,column,num_groups,group_values,group_labels=None):
    """
    Plots kaplan meire estimation of the survival function for an input column within a dataframe. If the column is numerical
    this function automatically discretizes the variable into 10-bin quantiles.
    args: Dataframe, column name to plot as a string, number of discrete groups to plot, which groups to plot, labels of the different
    categorical levels of a feature default is none
    Returns: None -- side effects are kaplan meire survival function plots.
    """
    df_temp = df.copy(deep=True)
    targets = ['default_time','status_time']
    if df[column].dtype == 'object':
        if num_groups == 2:
            df_0 =  df[df[column]==group_values[0]]
            df_1 =  df[df[column]==group_values[1]]

            kmf_0 = KaplanMeierFitter()
            kmf_1 = KaplanMeierFitter()

            kmf_0.fit(df_0['time'],df_0['default_time'],label=group_labels[0])
            kmf_1.fit(df_1['time'],df_1['default_time'],label=group_labels[1])

            kmf_0.plot_survival_function()
            kmf_1.plot_survival_function()

            plt.ylabel('Probability of survival (Probability of not defaulting)')
            plt.xlabel('Months')

            add_at_risk_counts(kmf_0,kmf_1)
            plt.title(f'Survival function for groups in {column}')
            plt.tight_layout()
            plt.show()
        elif num_groups == 3:
            df_0 =  df[df[column]==group_values[0]]
            df_1 =  df[df[column]==group_values[1]]
            df_2 =  df[df[column]==group_values[2]]

            kmf_0 = KaplanMeierFitter()
            kmf_1 = KaplanMeierFitter()
            kmf_2 = KaplanMeierFitter()

            kmf_0.fit(df_0['time'],df_0['default_time'],label=group_labels[0])
            kmf_1.fit(df_1['time'],df_1['default_time'],label=group_labels[1])
            kmf_2.fit(df_2['time'],df_2['default_time'],label=group_labels[2])

            kmf_0.plot_survival_function()
            kmf_1.plot_survival_function()
            kmf_2.plot_survival_function()

            plt.ylabel('Probability of survival (Probability of not defaulting)')
            plt.xlabel('Months')

            add_at_risk_counts(kmf_0,kmf_1,kmf_2)
            plt.title(f'Survival function for groups in {column}')
            plt.tight_layout()
            plt.show()
        elif num_groups == 4:
            df_0 =  df[df[column]==group_values[0]]
            df_1 =  df[df[column]==group_values[1]]
            df_2 =  df[df[column]==group_values[2]]
            df_3 =  df[df[column]==group_values[3]]

            kmf_0 = KaplanMeierFitter()
            kmf_1 = KaplanMeierFitter()
            kmf_2 = KaplanMeierFitter()
            kmf_3 = KaplanMeierFitter()

            kmf_0.fit(df_0['time'],df_0['default_time'],label=group_labels[0])
            kmf_1.fit(df_1['time'],df_1['default_time'],label=group_labels[1])
            kmf_2.fit(df_2['time'],df_2['default_time'],label=group_labels[2])
            kmf_3.fit(df_3['time'],df_3['default_time'],label=group_labels[3])

            kmf_0.plot_survival_function()
            kmf_1.plot_survival_function()
            kmf_2.plot_survival_function()
            kmf_3.plot_survival_function()

            plt.ylabel('Probability of survival (Probability of not defaulting)')
            plt.xlabel('Months')

            add_at_risk_counts(kmf_0,kmf_1,kmf_2,kmf_3)
            plt.title(f'Survival function for groups in {column}')
            plt.tight_layout()
            plt.show()

        elif num_groups == 5:
            df_0 =  df[df[column]==group_values[0]]
            df_1 =  df[df[column]==group_values[1]]
            df_2 =  df[df[column]==group_values[2]]
            df_3 =  df[df[column]==group_values[3]]
            df_4 =  df[df[column]==group_values[4]]

            kmf_0 = KaplanMeierFitter()
            kmf_1 = KaplanMeierFitter()
            kmf_2 = KaplanMeierFitter()
            kmf_3 = KaplanMeierFitter()
            kmf_4 = KaplanMeierFitter()

            kmf_0.fit(df_0['time'],df_0['default_time'],label=group_labels[0])
            kmf_1.fit(df_1['time'],df_1['default_time'],label=group_labels[1])
            kmf_2.fit(df_2['time'],df_2['default_time'],label=group_labels[2])
            kmf_3.fit(df_3['time'],df_3['default_time'],label=group_labels[3])
            kmf_4.fit(df_4['time'],df_4['default_time'],label=group_labels[4])

            kmf_0.plot_survival_function()
            kmf_1.plot_survival_function()
            kmf_2.plot_survival_function()
            kmf_3.plot_survival_function()
            kmf_4.plot_survival_function()

            plt.ylabel('Probability of survival (Probability of not defaulting)')
            plt.xlabel('Months')

            add_at_risk_counts(kmf_0,kmf_1,kmf_2,kmf_3,kmf_4)
            plt.title(f'Survival function for groups in {column}')
            plt.tight_layout()
            plt.show()

        else: 
            print('Must provide 2-5 groups')
    else:
        if any(column in var for var in targets):
            pass
        elif mortgage_df[column].isin([0,1]).all() == True:
            df_0 =  df[df[column]==0]
            df_1 =  df[df[column]==1]

            kmf_0 = KaplanMeierFitter()
            kmf_1 = KaplanMeierFitter()

            kmf_0.fit(df_0['time'],df_0['default_time'],label='No')
            kmf_1.fit(df_1['time'],df_1['default_time'],label='Yes')

            kmf_0.plot_survival_function()
            kmf_1.plot_survival_function()

            plt.ylabel('Probability of survival (Probability of not defaulting)')
            plt.xlabel('Months')

            add_at_risk_counts(kmf_0,kmf_1)
            plt.title(f'Survival function for groups in {column}')
            plt.tight_layout()
            plt.show()
        else:
            df_temp[column+'_discretized'] = pd.qcut(df[column],q=10,labels=False, duplicates='drop')
            # Check if col is binary
            
            if num_groups == 2:
                df_0 =  df_temp[df_temp[column+'_discretized']==group_values[0]]
                df_1 =  df_temp[df_temp[column+'_discretized']==group_values[1]]

                kmf_0 = KaplanMeierFitter()
                kmf_1 = KaplanMeierFitter()

                kmf_0.fit(df_0['time'],df_0['default_time'],label=str(group_values[0])+'Quantile')
                kmf_1.fit(df_1['time'],df_1['default_time'],label=str(group_values[1])+'Quantile')

                kmf_0.plot_survival_function()
                kmf_1.plot_survival_function()

                plt.ylabel('Probability of survival (Probability of not defaulting)')
                plt.xlabel('Months')

                add_at_risk_counts(kmf_0,kmf_1)
                plt.title(f'Survival function for groups in {column}_discretized')
                plt.tight_layout()
                plt.show()
                    
            elif num_groups == 3:
                df_0 =  df_temp[df_temp[column+'_discretized']==group_values[0]]
                df_1 =  df_temp[df_temp[column+'_discretized']==group_values[1]]
                df_2 =  df_temp[df_temp[column+'_discretized']==group_values[2]]

                kmf_0 = KaplanMeierFitter()
                kmf_1 = KaplanMeierFitter()
                kmf_2 = KaplanMeierFitter()

                kmf_0.fit(df_0['time'],df_0['default_time'],label=str(group_values[0])+'Quantile')
                kmf_1.fit(df_1['time'],df_1['default_time'],label=str(group_values[1])+'Quantile')
                kmf_2.fit(df_2['time'],df_2['default_time'],label=str(group_values[2])+'Quantile')

                kmf_0.plot_survival_function()
                kmf_1.plot_survival_function()
                kmf_2.plot_survival_function()

                plt.ylabel('Probability of survival (Probability of not defaulting)')
                plt.xlabel('Months')

                add_at_risk_counts(kmf_0,kmf_1,kmf_2)
                plt.title(f'Survival function for groups in {column}_discretized')
                plt.tight_layout()
                plt.show()
            elif num_groups == 4:
                df_0 =  df_temp[df_temp[column+'_discretized']==group_values[0]]
                df_1 =  df_temp[df_temp[column+'_discretized']==group_values[1]]
                df_2 =  df_temp[df_temp[column+'_discretized']==group_values[2]]
                df_3 =  df_temp[df_temp[column+'_discretized']==group_values[3]]

                kmf_0 = KaplanMeierFitter()
                kmf_1 = KaplanMeierFitter()
                kmf_2 = KaplanMeierFitter()
                kmf_3 = KaplanMeierFitter()

                kmf_0.fit(df_0['time'],df_0['default_time'],label=str(group_values[0])+'Quantile')
                kmf_1.fit(df_1['time'],df_1['default_time'],label=str(group_values[1])+'Quantile')
                kmf_2.fit(df_2['time'],df_2['default_time'],label=str(group_values[2])+'Quantile')
                kmf_3.fit(df_3['time'],df_3['default_time'],label=str(group_values[3])+'Quantile')

                kmf_0.plot_survival_function()
                kmf_1.plot_survival_function()
                kmf_2.plot_survival_function()
                kmf_3.plot_survival_function()

                plt.ylabel('Probability of survival (Probability of not defaulting)')
                plt.xlabel('Months')

                add_at_risk_counts(kmf_0,kmf_1,kmf_2,kmf_3)
                plt.title(f'Survival function for groups in {column}_discretized')
                plt.tight_layout()
                plt.show()

            elif num_groups == 5:
                df_0 =  df_temp[df_temp[column+'_discretized']==group_values[0]]
                df_1 =  df_temp[df_temp[column+'_discretized']==group_values[1]]
                df_2 =  df_temp[df_temp[column+'_discretized']==group_values[2]]
                df_3 =  df_temp[df_temp[column+'_discretized']==group_values[3]]
                df_4 =  df_temp[df_temp[column+'_discretized']==group_values[4]]

                kmf_0 = KaplanMeierFitter()
                kmf_1 = KaplanMeierFitter()
                kmf_2 = KaplanMeierFitter()
                kmf_3 = KaplanMeierFitter()
                kmf_4 = KaplanMeierFitter()

                kmf_0.fit(df_0['time'],df_0['default_time'],label=str(group_values[0])+'Quantile')
                kmf_1.fit(df_1['time'],df_1['default_time'],label=str(group_values[1])+'Quantile')
                kmf_2.fit(df_2['time'],df_2['default_time'],label=str(group_values[2])+'Quantile')
                kmf_3.fit(df_3['time'],df_3['default_time'],label=str(group_values[3])+'Quantile')
                kmf_4.fit(df_3['time'],df_3['default_time'],label=str(group_values[4])+'Quantile')

                kmf_0.plot_survival_function()
                kmf_1.plot_survival_function()
                kmf_2.plot_survival_function()
                kmf_3.plot_survival_function()
                kmf_4.plot_survival_function()

                plt.ylabel('Probability of survival (Probability of not defaulting)')
                plt.xlabel('Months')

                add_at_risk_counts(kmf_0,kmf_1,kmf_2,kmf_3,kmf_4)
                plt.title(f'Survival function for groups in {column}_discretized')
                plt.tight_layout()
                plt.show()

            else: 
                print('Must provide 2-5 groups')
        
```

### CoxPhitter 

* **Cox's proportional hazard model's interpretation:**
Given the censored data we have, ordinary linear regression isn't an option. Luckily, the `lifelines` library offers Cox's proportional hazard model, allowing for a regression of the covariates of interest against survival durations. Cox's model is considered semi-parametric, as it uses a baseline hazard function to estimate the hazard rate for all covariates when their values are zero. That being said, what is of interest is not the accuracy of the hazard rate exactly, but how it changes in relation to the covariates with time. Additionally, given the data we have is not static and the covariates fluctuate over time, lifelines has the `CoxTimeVaryingFitter` to support this sort of data. An instance of the time varying Cox's model is initiated in the cell just below this one. Of course only the non-static features are fed to the model in this case. 

In the summary table output found using the `summary` attribute, we can observe the coefficients which are made more interpretable by transformation using the exponential function. Therefore, I will remark on the transformed coefficients only. They can be interpreted as follows: A coefficient below 1.0 such as that of the `FICO_orig_time` covariate, indicates that a one unit increase of this covariate results in a decrease of the hazard rate of 1%. As mentioned previously, the hazard rate reflects the inverse of the probability of survival, so the likelihood of default essentially. We can see in the summary table that `gdp_time` decreases the hazard rate by the highest magnitude when it is increased (28.5%), while `interest_rate_time` shows the largest increase to the hazard rate as it increases monotonically (28%). This makes complete intuitive sense, as when the interest rate increases it is more likely that the borrower defaults. Additionally, all the coefficient's P values are signficant, meaing we can confidently reject the null hypothesis that these covariates don't actually have any impact on time-to-event. 

```python
# Include only time varying variables
mortgage_cp = mortgage_df[["id","orig_time","time","default_time", "hpi_time", "gdp_time", "uer_time", "balance_time",
"interest_rate_time", "FICO_orig_time", "LTV_time"]]

cph_tv = CoxTimeVaryingFitter()

cph_tv.fit(mortgage_cp,id_col='id',event_col='default_time', start_col='orig_time',stop_col='time',show_progress=True)
cph_tv.summary

```

```python
cph_tv.plot()
```

* **Cox's proportional hazard model's plot interpretation:** 

The plot in the cell block graphically depicts the insights discussed previously. The coefficients are back-transformed by taking the logarithm of the `exp(coef)` values for the summary table. We can see those covariates which increase the hazard rate the most to the right of the zero midpoint and those that do the opposite farthest to the left. We can also see the 95% confidence interval for the coefficient values, with those having the largest intervals showing the longest whiskers.
